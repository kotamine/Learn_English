```python
# Example of srt file:
# 1
# 00:00:37,364 --> 00:00:41,240
# Hey, let's go
# Hey, let's go 

# 2
# 00:00:41,396 --> 00:00:44,732
#  I'm happy as can be 

from datetime import datetime
import re

def read_srt_lines_as_dict(f): 
    lines = f.readlines()
    
    ts_type = type(datetime.strptime('00:00:00', '%H:%M:%S').time())

    # target dictionary to fill in 
    dict1 = ({})

    # initialize data
    num = 0
    timing_beg = datetime.strptime('00:00:00', '%H:%M:%S').time()
    timing_end = datetime.strptime('00:00:00', '%H:%M:%S').time()
    string1 = ''

    for n, line in enumerate(lines):      
        line = str.replace(line,'\n','')
        if str.isdigit(line): 
            num = line
        elif len(re.findall('^[0-9][0-9]:[0-9][0-9]:[0-9][0-9]',line))>0:
            timing_beg = datetime.strptime(str.split(line,',')[0], '%H:%M:%S').time()
            timing_end = datetime.strptime(str.split(str.split(line,'--> ')[1],',')[0], '%H:%M:%S').time()
        elif len(line) == 0: 
            # empty line = the end of each set of subtitles 
            print({num: [timing_beg, timing_end, string1]})
            dict1.update({num: [timing_beg, timing_end, string1]})
            # reset 
            num = 0
            timing_beg = datetime.strptime('00:00:00', '%H:%M:%S').time()
            timing_end = datetime.strptime('00:00:00', '%H:%M:%S').time()
            string1 = ''
        else:
            string1 += line + ' ' 

    return pd.DataFrame.from_dict(dict1, orient='index')
```

```python

def read_txt_lines_as_dict(f): 
    lines = f.readlines()
    
    # target dictionary to fill in 
    dict1 = ({})
    
    for n, line in enumerate(lines):      
        line = str.replace(line,'\n','')
        dict1.update({n: [line]})
    return pd.DataFrame.from_dict(dict1, orient='index')
```

```python
import requests
from bs4 import BeautifulSoup

def scrape_ejje(word):
    url = "https://ejje.weblio.jp/content/" + word
    source = requests.get(url)
    source.encoding = source.apparent_encoding
    data = BeautifulSoup(source.text, "html.parser")
    explanation_list = data.select("span.content-explanation.ej")
    meanings = ''.join([explanation_list[idx].text for idx, txt in enumerate(explanation_list)])   
    return meanings
```
